```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# set a style
sns.set(style="whitegrid", palette="pastel")

# Load dataset
DATA_PATH = Path("../data/raw/world_happiness.csv")
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
df.info()
df.describe()

# missing values percentage
df.isna().mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(9,6))
sns.heatmap(
    df[[
        "GDP per capita",
        "Social support",
        "Healthy life expectancy",
        "Freedom to make life choices",
        "Generosity",
        "Perceptions of corruption",
        "Score"
    ]].corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Feature Correlation with Happiness Score")
plt.show()
```

```python
sns.pairplot(
    df,
    vars=[
        "GDP per capita",
        "Social support",
        "Healthy life expectancy",
        "Freedom to make life choices",
        "Generosity",
        "Perceptions of corruption",
        "Score"
    ],
    corner=True
)
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

```python
features = [
    "GDP per capita",
    "Social support",
    "Healthy life expectancy",
    "Freedom to make life choices",
    "Generosity",
    "Perceptions of corruption"
]

for f in features:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=df[f], y=df["Score"])
    plt.title(f"{f} vs Happiness Score")
    plt.show()
```

```python
fig_dir = Path("../reports/figures")
fig_dir.mkdir(parents=True, exist_ok=True)

plt.figure(figsize=(9,6))
sns.heatmap(df[features + ["Score"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation with Happiness Score")
plt.tight_layout()
plt.savefig(fig_dir / "correlation_heatmap.png")
plt.close()

print("✅ Saved correlation heatmap to reports/figures/")
```
